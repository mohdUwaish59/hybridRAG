```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json
import os

# Load JSON files safely
def load_json(filename):
    if not os.path.exists(filename):
        print(f"Error: File '{filename}' not found.")
        return None
    try:
        with open(filename, 'r') as f:
            return json.load(f)
    except json.JSONDecodeError:
        print(f"Error: Unable to decode '{filename}'. Ensure it is a valid JSON file.")
        return None

# Load the data from JSON files
hybrid_data = load_json("hybridRAG/bleu_score_results.json")
baseline_data = load_json("baselineRAG/baseline_bleu_score_results.json")

# Proceed only if data is loaded successfully
if hybrid_data and baseline_data:
    # Create DataFrames
    hybrid_df = pd.DataFrame(hybrid_data)
    baseline_df = pd.DataFrame(baseline_data)

    # Add system identifier
    hybrid_df['System'] = 'Hybrid'
    baseline_df['System'] = 'Baseline'

    # Combine the data
    combined_df = pd.concat([hybrid_df, baseline_df], ignore_index=True)

    # Define category order
    category_order = ['Explicit Fact', 'Implicit Fact', 'Hidden Rationale', 'Interpretable Rationale']
    
    # Calculate average score for each category per system
    pivot_data = combined_df.groupby(['Predicted_Category', 'System'])['Score'].mean().unstack()

    # Reorder categories
    pivot_data = pivot_data.reindex(category_order)

    # Plot the line chart
    plt.figure(figsize=(10, 6))

    # Plot lines for Hybrid and Baseline
    plt.plot(pivot_data.index, pivot_data['Hybrid'], marker='o', linestyle='-', color='blue', label='Hybrid System')
    plt.plot(pivot_data.index, pivot_data['Baseline'], marker='s', linestyle='--', color='orange', label='Baseline System')

    # Add labels and title
    plt.xlabel('Query Category')
    plt.ylabel('Average BLEU Score')
    plt.title('BLEU Score Comparison: HybridRAG vs Baseline System')
    plt.xticks(rotation=45)
    plt.legend()

    # Show the grid for better readability
    plt.grid(True, linestyle='--', alpha=0.7)

    # Save the plot
    plt.savefig("bleu_score_comparison.png", dpi=300)

    # Show the plot
    plt.show()
```
